```python
import wandb
import pandas as pd
import os

# 1️⃣ Run starten
run = wandb.init(project="nyc_airbnb", group="eda", save_code=True)

# 2️⃣ Artifact herunterladen (gibt einen Ordner zurück)
artifact = wandb.use_artifact("sample.csv:latest")
artifact_dir = artifact.download()  # Windows-kompatibel!

# 3️⃣ Pfad zur Datei zusammensetzen
# Prüfe, wie die Datei tatsächlich heißt:
print(os.listdir(artifact_dir))

# Angenommen, die Datei heißt 'sample.csv'
local_path = os.path.join(artifact_dir, "sample.csv")
```
